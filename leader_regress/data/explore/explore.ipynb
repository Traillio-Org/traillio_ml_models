```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_leetcode = pd.read_json("../leetcode_data.json")
df_leetcode.head()
df_leetcode = df_leetcode.dropna(how='all')
```

```python
df_codeforces = pd.read_json("../codeforces_data.json")
df_codeforces.head()
```

```python
leetcode_columns = df_leetcode.columns
leetcode_columns
```

```python
df_leetcode = df_leetcode.drop(columns=[leetcode_columns[2], leetcode_columns[4], leetcode_columns[6], leetcode_columns[8], leetcode_columns[9], leetcode_columns[10], leetcode_columns[11], leetcode_columns[13], leetcode_columns[14]])
df_leetcode.head()
```

```python
def sum_of_ques(items):
    sum = 0
    for item in items:
        sum += item[1]
    return sum
```

```python
def filter_last_n_keys_dict(d, n=7):
    if isinstance(d, dict):
        return sum_of_ques(list(d.items())[-n:])
df_leetcode["submissionCalendar"] = df_leetcode["submissionCalendar"].apply(lambda x: filter_last_n_keys_dict(x, 7))
```

```python
total_list = df_leetcode["totalSubmissions"]
final_list = []
for item in total_list:
    # print(type(item))
    if(isinstance(item, list)):
        if isinstance(
            item[0], dict
        ):  # Ensure it's a dictionary before attempting to access keys
            final_list.append(round(item[0]["count"] / item[0]["submissions"], 2))
        else:
            print(f"Unexpected value: {item}")  # Print if it's not a dictionary
final_list
```

```python
final_list = np.array(final_list)
```

```python
final_list
```

```python
df_leetcode["totalSubmissions"] = final_list
```

```python
df_codeforces = df_codeforces.iloc[:, codeforces_columns]
df_codeforces = df_codeforces.drop(columns=["username"])
df_codeforces.head()
```
